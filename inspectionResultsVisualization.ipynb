```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv('Rodent_Inspection_20250428.csv')

df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='RESULT', order=df['RESULT'].value_counts().index, palette='dark')
plt.title('Distribution of Rodent Inspection Results')
plt.xlabel('Inspection Result')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
